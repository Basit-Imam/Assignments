### 1.1 Write a Python Program to implement your own myreduce() function which works exactly like Python's built-in function reduce()

```python
#Sum of n numbers using user defined myreduce function.

def myreduce(num_list):
    sum = 0
    for i in num_list:
        sum +=i
        
    return sum

ls = []
n = int(input("Enter the number of elements of list "))
for i in range(0,n):
    element = int(input())
    ls.append(element)
print(ls)
print ("Sum of the given list using custom reduce function is "   + str(myreduce(ls)))
```

### 1.2 Write a Python program to implement your own myfilter() function which works exactly like Python's built-in function filter()

```python
# Filter the array, and return a new array with only the values equal to or above 18:

def myfilter(num_list):
    new_list = []
    for age in num_list:
        if age >= 18:
            new_list.append(age)
        else :
            continue
    return new_list

n=int(input("Please insert the number: "))
ls = []
for i in range(0, n):
    element = int(input())
    ls.append(element)
    
print(ls)
print ("New array having values only greater than or equal to 18 "   + str(myfilter(ls)))
```

```python
word = list('xyz')
word_list = [n*i for i in range(1,5) for n in word]
print(word_list)
```

```python
num = [2, 3, 4]
new_num = [[n+i] for n in num for i in range(0,3)]
print(new_num)
```

```python
num = [2, 3, 4, 5]
new_num = [[n+i for i in range(0,4)] for n in num]
print(new_num)
```

```python
num = [1, 2, 3]
new_num = [(a, b) for b in num for a in num]
print(new_num)
```
